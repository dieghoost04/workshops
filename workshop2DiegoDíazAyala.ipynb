```python
# clear the plot space:
plt.clf()
# Define the probability type of function as uniform (probability space)
P = Uniform(0,40)
# Define X as a random variable of the type P (uniform)
X =RV(P)
# Simulate the values of X and save the in the x array: 
x=X.sim(10000)
x.plot()
plt.show()
```

The variance is calculated as the mean of the squared differences between each value and the mean. Since all values ââwithin the interval [a,b] are equally likely, we can use the midpoint as the most representative value:


$$
\overline{x} = \frac{(a+b)}{2}\\
Var(x) = \int(x-\overline{x})^{2}*f(x) dx
$$

$$Var(x) = \frac{(b-a)^2}{12}$$
